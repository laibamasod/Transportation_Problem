```python
from math import radians, sin, cos, sqrt, atan2
import networkx as nx


# Create a graph
G = nx.Graph()

# Add nodes (Markaz) to the graph
for markaz, coord in markaz_locations.items():
    G.add_node(markaz, pos=coord)
    
    
# Function to calculate distance between two points using Haversine formula
def calculate_distance(lat1, lon1, lat2, lon2):
    # Convert latitude and longitude from degrees to radians
    lat1_rad = radians(lat1)
    lon1_rad = radians(lon1)
    lat2_rad = radians(lat2)
    lon2_rad = radians(lon2)
    
    # Radius of the Earth in kilometers
    radius = 6371.0
    
    # Haversine formula
    dlon = lon2_rad - lon1_rad
    dlat = lat2_rad - lat1_rad
    a = sin(dlat / 2)**2 + cos(lat1_rad) * cos(lat2_rad) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))
    distance = radius * c
    
    return distance

# Markaz locations with their coordinates
markaz_locations = {
    'I-8 Markaz': (33.667946114614175, 73.07502035923504),
    'I-10 Markaz': (33.647021286426785, 73.03740393992595),
    'F-6 Markaz': (33.72987997328698, 73.07605232898078),
    'F-7 Markaz': (33.719506265052615, 73.05489404445248),
    'F-8 Markaz': (33.71171807077601, 73.0396843017092),
    'F-11 Markaz': (33.68532436267236, 72.98717896050887),
    'G-9 Markaz': (33.68727719709082, 73.0346779475553),
    'G-8 Markaz': (33.69713779180326, 73.05048694129904)
}

# Calculate distances between Markaz
markaz_distances = {}
for m1, coord1 in markaz_locations.items():
    markaz_distances[m1] = {}
    for m2, coord2 in markaz_locations.items():
        if m1 != m2:
            distance = calculate_distance(coord1[0], coord1[1], coord2[0], coord2[1])
            markaz_distances[m1][m2] = round(distance,2)
            # Add edges with distances as weights
            G.add_edge(m1, m2, weight=round(distance,2))

# Specify the connections to be removed
connections_to_remove = [
    ('I-8 Markaz', 'F-6 Markaz'),
    ('I-8 Markaz', 'F-8 Markaz'),
    ('I-8 Markaz', 'F-7 Markaz'),
    ('F-6 Markaz', 'F-8 Markaz'),
    ('F-6 Markaz', 'I-10 Markaz'),
    ('F-6 Markaz', 'G-9 Markaz'),
    ('F-8 Markaz', 'I-10 Markaz'),
    ('F-11 Markaz', 'F-7 Markaz'),
    ('F-11 Markaz', 'F-6 Markaz'),
    ('F-11 Markaz', 'I-8 Markaz'),
    ('F-7 Markaz', 'I-10 Markaz'),
    ('F-7 Markaz', 'G-9 Markaz'),
    ('G-8 Markaz', 'F-11 Markaz'),
    ('G-8 Markaz', 'I-10 Markaz')
]

# Update the distances to simulate unavailable connections
for conn in connections_to_remove:
    m1, m2 = conn
    if m1 in markaz_distances and m2 in markaz_distances[m1]:
        markaz_distances[m1][m2] = float('inf')  # Set distance as infinite to simulate no connection
        markaz_distances[m2][m1] = float('inf')
        if G.has_edge(m1, m2):
            G.remove_edge(m1, m2)  # Remove the edge from the graph
        # Update the reverse direction as well
        if G.has_edge(m2, m1):
            G.remove_edge(m2, m1)  # Remove the reverse edge from the graph

# Display the distances between Markaz
for m1, distances in markaz_distances.items():
    print(f"Distances from {m1}:")
    for m2, distance in distances.items():
        if distance != float('inf'):
            print(f"To {m2}: {distance:.2f} km")
        else:
            print(f"To {m2}: Not accessible")
    print()

```

```python
# Function for A* Search Algorithm
import heapq

def a_star_search(graph, start, goal):
    open_set = [(0, start)]  # Priority queue for nodes to be evaluated
    came_from = {}  # Keeps track of the path
    g_score_as = {node: float('inf') for node in graph}  # Cost from start along the best-known path
    g_score_as[start] = 0
    f_score_as = {node: float('inf') for node in graph}  # Estimated total cost from start to goal through node
    f_score_as[start] = heuristic(start, goal)  # Heuristic function for initial node
    explored_nodes_as = 0  # Store the total count of explored nodes

    while open_set:
        current = heapq.heappop(open_set)[1]
        explored_nodes_as += 1  # Increment the explored node count

        if current == goal:
            return g_score_as[current], reconstruct_path_as(came_from, current), explored_nodes_as

        for neighbor in graph[current]:
            tentative_g_score = g_score_as[current] + graph[current][neighbor]['weight']

            if tentative_g_score < g_score_as[neighbor]:
                came_from[neighbor] = current
                g_score_as[neighbor] = tentative_g_score
                f_score_as[neighbor] = tentative_g_score + heuristic(neighbor, goal)
                heapq.heappush(open_set, (f_score_as[neighbor], neighbor))

    return float('inf'), None, explored_nodes_as  # If no path found


def reconstruct_path_as(came_from, current):
    path = [current]
    while current in came_from:
        current = came_from[current]
        path.insert(0, current)
    return path


def manhattan_distance(a, b):
    return abs(a[0] - b[0]) + abs(a[1] - b[1])

def heuristic(a, b):
    # Assuming a and b are keys in the markaz_locations dictionary
    coord_a = markaz_locations[a]
    coord_b = markaz_locations[b]
    return manhattan_distance(coord_a, coord_b)

```

```python
# Function for Uniform Cost Search Algorithm

import heapq

def uniform_cost_search(graph, start, goal):
    open_set = [(0, start)]  # Priority queue for nodes to be evaluated
    came_from = {}  # Keeps track of the path
    g_score = {node: float('inf') for node in graph}  # Cost from start along the best-known path
    g_score[start] = 0
    explored_nodes = 0  # Store the total count of explored nodes

    while open_set:
        current_cost, current = heapq.heappop(open_set)
        explored_nodes += 1  # Increment the explored node count

        if current == goal:
            return g_score[current], reconstruct_path(came_from, current), explored_nodes

        for neighbor in graph[current]:
            tentative_g_score = g_score[current] + graph[current][neighbor]['weight']

            if tentative_g_score < g_score[neighbor]:
                came_from[neighbor] = current
                g_score[neighbor] = tentative_g_score
                heapq.heappush(open_set, (tentative_g_score, neighbor))

    return None, None, explored_nodes  # If no path found
```

```python
# Choose source and destination randomly from the Markaz locations
source_markaz = 'F-11 Markaz'
destination_markaz = 'F-6 Markaz'
```

```python
# Apply A* Search

import time

start_time = time.time()  # Record start time
astar_cost, astar_path, astar_explored = a_star_search(G, source_markaz, destination_markaz)
print("A* Search - Cost:", astar_cost)
print("A* Search - Path:", astar_path)
print("A* Search - Explored Nodes Count:", astar_explored)

end_time = time.time()  # Record end time
execution_time = end_time - start_time  # Calculate execution time
print(f"Time Taken for A* Search: {execution_time}")
```

```python
# Apply Uniform Cost Search

import time

start_time = time.time()  # Record start time
uc_cost, uc_path, uc_explored = uniform_cost_search(G, source_markaz, destination_markaz)
print("Uniform Cost Search - Cost:", uc_cost)
print("Uniform Cost Search - Path:", uc_path)
print("Uniform Cost Search - Explored Nodes Count:", uc_explored)

end_time = time.time()  # Record end time
execution_time = end_time - start_time  # Calculate execution time
print(f"Time Taken for Uniform Cost Search: {execution_time}")
```

Both seem like efficient algorithms, However, the A* takes a bit less time compared to UCS, but the difference is negligible. Introducing more nodes may help us distinguish better between the two.
